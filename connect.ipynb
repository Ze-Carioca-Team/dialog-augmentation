```python
import pandas as pd
import json
from pprint import pprint
```

```python
data = json.load(open("final.json"))
```

# Hard cases

```python
issues = []
delim = "Motivo da contestação:*"

for dialog in data:
  issue = dialog["turns"][0]['text'].find(delim)
  if issue > 0:
    issues.append(dialog["turns"][0]['text'][issue+len(delim):].split("*")[0].strip())
```

```python
from collections import Counter
import pprint
pprint.pprint(Counter(issues))
```

```python
filtered = []
delim = "Foi cobrado um valor mais alto"
for dialog in data:
  issue = dialog["turns"][0]['text'].find(delim)
  if issue > 0:
    filtered.append(dialog)
```

```python
import random
import torch
import nlpaug.augmenter.word as naw

#aug = naw.SynonymAug(aug_src='wordnet', lang='por')
aug = naw.ContextualWordEmbsAug(model_path='bert-base-multilingual-uncased', aug_p=0.1)

def expansion():
    samples = [
        "Foi cobrado um valor mais alto.",
        "Fui cobrado incorretamente.",
        "Cobraram a mais na fatura.",
        "Me cobraram a mais."]
    greetings = [
        "Olá.",
        "Bom dia.",
        "Boa tarde.",
        "Boa noite.",
        "Oi."]
    resp = random.choice(samples)
    greet = random.random()
    intent = '[problema_fatura]'
    if greet > 0.5:
        resp = random.choice(greetings)+" "+resp
        intent = '[greeting] '+intent
    return resp, intent
```

```python
parsed = []
counter = 0
for dialog in filtered:
    new_dialog = {"id":dialog["id"], "turns":[]}
    for index, turn in enumerate(dialog["turns"]):
        if index == 0:
            new_turn = {}
            new_turn["type"] = turn["type"]
            exp = expansion()
            new_turn["text"] = exp[0]
            new_turn["intent"] = exp[1]
            new_dialog["turns"].append(new_turn)
        #elif index ==:
        elif new_dialog["turns"][-1]["type"] == turn["type"]:
            new_turn["text"] += " "+turn["text"]
        else:
            new_turn = {}
            new_turn["type"] = turn["type"]
            new_turn["text"] = turn["text"]
            new_dialog["turns"].append(new_turn)
    parsed.append(new_dialog)
    counter += 1
print(counter)
```

```python
del filtered
```

```python
random.sample(parsed, 10)
```
